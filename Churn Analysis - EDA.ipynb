```python
telcom_data = telcom_base_data.copy()
```

```python
telcom_data.TotalCharges = pd.to_numeric(telcom_data.TotalCharges, errors = 'coerce')
telcom_data.isnull().sum()
```

```python
telcom_data.loc[telcom_data['TotalCharges'].isnull() == True]
```

```python
telcom_data.dropna(how = 'any', inplace = True)
```

```python
print(telcom_data['tenure'].max())
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

telcom_data['tenure_group'] = pd.cut(telcom_data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
telcom_data['tenure_group'].value_counts()
```

```python
#drop column customerID and tenure
telcom_data.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
telcom_data.head()
```

```python
for i, predictor in enumerate(telcom_data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=telcom_data, x=predictor, hue='Churn')
```

```python
telcom_data['Churn'] = np.where(telcom_data.Churn == 'Yes',1,0)
```

```python
telcom_data.head()
```

```python
telcom_data_dummies = pd.get_dummies(telcom_data)
telcom_data_dummies.head()
```

```python
sns.lmplot(data=telcom_data_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

```python
Mth = sns.kdeplot(
    telcom_data_dummies.MonthlyCharges[telcom_data_dummies["Churn"] == 0],
    color="red",
    fill=True,      # âœ” replacement for shade
    alpha=0.5,      # âœ” transparency so both curves are visible
    common_norm=False
)
sns.kdeplot(
    telcom_data_dummies.MonthlyCharges[telcom_data_dummies["Churn"] == 1],
    ax=Mth,
    color="blue",
    fill=True,
    alpha=0.5,
    common_norm=False
)
Mth.legend(["No Churn", "Churn"], loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
plt.show()

```

```python
Tot = sns.kdeplot(telcom_data_dummies.TotalCharges[(telcom_data_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Tot = sns.kdeplot(telcom_data_dummies.TotalCharges[(telcom_data_dummies["Churn"] == 1) ],
                ax =Tot, color="Blue", shade= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
uniplot(new_df1_target1,col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```
